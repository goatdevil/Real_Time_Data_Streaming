```python
import json


try:
    with open('data.json', 'r') as f:
        data=json.load(f)
except:
    data = json.loads('[{"nom": "Alice", "age": 30, "ville": "Paris"},{"nom": "Bob", "age": 25, "ville": "Lyon"},{"nom": "Charlie", "age": 35, "ville": "Marseille"}]')

for line in data:
    if line['nom']=='Bob':
        print(line['age'])
    elif line['nom']=='Alice':
        line['age']=31

new={"nom": "Daniel", "age": 40, "ville": "Nice"}
data.append(new)
print(data)

formatted_json=json.dumps(data, indent=4)
print(formatted_json)


with open('updated_data.json', 'w') as f:
    json.dump(data, f)
```
